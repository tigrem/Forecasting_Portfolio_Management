```python
import os
import yfinance as yf
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# Define the assets and date range
assets = ['TSLA', 'BND', 'SPY']
start_date = '2015-07-01'
end_date = '2025-07-31'

# --- Add this code to create the directories if they don't exist ---
raw_data_dir = 'data/row_data'
processed_data_dir = 'data/processed'

os.makedirs(raw_data_dir, exist_ok=True)
os.makedirs(processed_data_dir, exist_ok=True)
# -------------------------------------------------------------------

# Fetch the data
data = yf.download(assets, start=start_date, end=end_date)

# Display the first few rows to see the structure
print(data.head())

# Extract 'Close' prices for each asset (now contains adjusted prices)
adj_close_data = data['Close']

# Save the raw data to the `data/row_data` folder
# The `raw_data_dir` variable now holds the correct path
adj_close_data.to_csv(os.path.join(raw_data_dir, 'financial_data_raw.csv'))

# Display the first few rows of the adjusted close data
print(adj_close_data.head())
```

```python
# Check for basic statistics
print("\nDescriptive Statistics:")
print(adj_close_data.describe())

# Check for missing values
print("\nMissing values per asset:")
print(adj_close_data.isnull().sum())

# Handle missing values (if any). YFinance data is generally clean.
# For this dataset, we can use forward-fill if a day is missing
adj_close_data.fillna(method='ffill', inplace=True)

# Verify no more missing values
print("\nMissing values after handling:")
print(adj_close_data.isnull().sum())

# Check data types
print("\nData types:")
print(adj_close_data.info())

# No scaling is needed for this initial analysis, as we are working with raw prices
# and daily returns, not complex machine learning models that require it.
```

```python
# Use a simple Z-score approach for outlier detection in returns
# Z-score = (return - mean_return) / std_dev_of_returns

for asset in assets:
    returns_mean = daily_returns[asset].mean()
    returns_std = daily_returns[asset].std()
    
    # Identify returns with a Z-score > 3 or < -3 as outliers
    outliers = daily_returns[
        (daily_returns[asset] > returns_mean + 3 * returns_std) |
        (daily_returns[asset] < returns_mean - 3 * returns_std)
    ][asset]
    
    print(f"\nOutliers for {asset}:")
    print(outliers)
```
