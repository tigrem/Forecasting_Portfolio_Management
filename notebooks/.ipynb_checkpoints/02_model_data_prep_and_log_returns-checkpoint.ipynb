```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- Configuration ---
RAW_DATA_PATH = '../data/BrentOilPrices.csv'
PROCESSED_DATA_DIR = '../data/processed/'
PROCESSED_LOG_RETURNS_PATH = os.path.join(PROCESSED_DATA_DIR, 'brent_oil_log_returns.csv')

# Ensure processed data directory exists
os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)

print(f"Loading raw data from: {RAW_DATA_PATH}")

# --- 1. Load Brent Oil Price Data ---
try:
    df_raw = pd.read_csv(RAW_DATA_PATH)

    # Explicitly convert the 'Date' column to datetime objects
    df_raw['Date'] = pd.to_datetime(df_raw['Date'], format='%d-%b-%y', errors='coerce')

    # Drop rows where 'Date' could not be parsed
    df_raw = df_raw.dropna(subset=['Date'])

    # Ensure 'Price' is numeric, coercing errors to NaN and then dropping them
    df_raw['Price'] = pd.to_numeric(df_raw['Price'], errors='coerce')
    df_raw = df_raw.dropna(subset=['Price'])

    # --- 2. Set 'Date' as the index and sort ---
    df_raw = df_raw.set_index('Date').sort_index()

    print("Raw Data Head:")
    print(df_raw.head())
    print("\nRaw Data Info:")
    df_raw.info()

except FileNotFoundError:
    print(f"Error: {RAW_DATA_PATH} not found. Please ensure the file is in the correct location.")
    # Create a dummy DataFrame for demonstration if the file is missing
    dates = pd.date_range(start='1987-05-20', periods=1000, freq='D')
    prices = np.random.rand(1000) * 50 + 20  # Dummy prices
    df_raw = pd.DataFrame({'Price': prices}, index=dates)
    print("\nUsing dummy data for demonstration.")

except Exception as e:
    print(f"An unexpected error occurred during data loading: {e}")
    # Create a dummy DataFrame for demonstration if an error occurs
    dates = pd.date_range(start='1987-05-20', periods=1000, freq='D')
    prices = np.random.rand(1000) * 50 + 20  # Dummy prices
    df_raw = pd.DataFrame({'Price': prices}, index=dates)
    print("\nUsing dummy data for demonstration due to error.")

# --- 3. Calculate Log Returns ---
df_log_returns = np.log(df_raw['Price']).diff().dropna().to_frame(name='Log_Return')

print("\nLog Returns Data Head:")
print(df_log_returns.head())
print("\nLog Returns Data Info:")
df_log_returns.info()

# --- 4. Visualize Raw Prices and Log Returns ---
plt.figure(figsize=(15, 10))

# Plot Raw Prices
plt.subplot(2, 1, 1)
plt.plot(df_raw.index, df_raw['Price'], label='Brent Oil Price (USD/barrel)')
plt.title('Brent Oil Prices (1987-2022)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()

# Plot Log Returns
plt.subplot(2, 1, 2)
plt.plot(df_log_returns.index, df_log_returns, label='Daily Log Returns')
plt.title('Daily Log Returns of Brent Oil Prices')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# --- 5. Save Processed Data ---
df_log_returns.to_csv(PROCESSED_LOG_RETURNS_PATH)
print(f"\nLog returns saved to: {PROCESSED_LOG_RETURNS_PATH}")
```
